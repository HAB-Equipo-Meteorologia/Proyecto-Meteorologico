```python
import pandas as pd
import joblib
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../../data/ml/historical_clean_sample.csv', parse_dates=['fecha'])
```

```python
df = df.set_index('fecha')
```

```python
ignore_cols = ['year', 'fecha_sin', 'fecha_cos']
```

```python
minmax_scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
minmax_scaler.fit_transform(df.drop(columns=ignore_cols))
joblib.dump(minmax_scaler, '../../data/ml/minmax_scaler.joblib')
```

```python
def scale_df(df, scaler):

    df_scaled = df.drop(columns=ignore_cols)
    
    scaled_data = scaler.transform(df_scaled)
    
    df_scaled = pd.DataFrame(scaled_data, columns=df_scaled.columns)
    df_scaled.set_index(df.index, inplace=True)
    df_scaled[ignore_cols] = df[ignore_cols]

    df_scaled = df_scaled.round(4)

    return df_scaled
```

```python
train_df_scaled = scale_df(df, minmax_scaler)
train_df_scaled.head()
```

```python
train_df_scaled.to_csv('../../data/ml/historical_clean_scaled.csv')
```
